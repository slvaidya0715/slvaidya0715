## Chart - 4

### -Percentage Availability by Room Type

```python
room_type_frequency = airbnb_df['room_type'].value_counts()

total_days_in_year = 365
availability_by_room_type = airbnb_df.groupby('room_type')['availability_365'].sum()
percentage_availability = (availability_by_room_type / (room_type_frequency * total_days_in_year)) * 100
```

```python
#the graph
plt.figure(figsize=(8, 6))
percentage_availability.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Percentage Availability by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Percentage Availability')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 1. Why did you pick the specific chart?

The bar chart's effectiveness lies in its ability to represent numerical values with the height of the bars, making it intuitive for viewers to understand and compare the values among different categories.

### 2. What is/are the insight(s) found from the chart?

**Lower Demand for Shared Rooms:** The fact that "shared room" has the highest availability percentage indicates that there might be lower demand for this type of accommodation compared to private rooms or entire homes/apartments. Guests might prefer more private and exclusive accommodations, leading to higher availability of shared rooms.

## Chart- 15

### -Correlation Heatmap

```python
# Select numerical features for the correlation matrix
numerical_features = airbnb_df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']]

# Create the correlation matrix
correlation_matrix = numerical_features.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix between Numerical Features", fontsize=16)
plt.show()
```

### 1. Why did you pick the specific chart?

Heatmaps are an effective way to visualize the correlation between numerical features in a dataset. In this case, the heatmap provides a clear and concise representation of the correlation coefficients between different numerical features in the airbnb_df dataset.

### 2. What is/are the insight(s) found from the chart?

From the heatmap, we can observe the correlation strength and direction between pairs of numerical features. The color intensity and the corresponding values in the cells indicate the level of correlation, with positive values denoting a positive correlation, negative values indicating a negative correlation, and values close to zero representing weak or no correlation. For example, we can see that there is a positive correlation between the "number_of_reviews" and "reviews_per_month" features, which is expected, as listings with more reviews per month will generally have a higher total number of reviews. Additionally, there seems to be a weak negative correlation between "availability_365" and "number_of_reviews," suggesting that listings with higher availability tend to receive fewer reviews.
